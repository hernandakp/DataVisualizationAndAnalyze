# Produksi Daging Kambing Berdasarkan Kabupaten/Kota di Jawa Barat Tahun 2016-2021

**Metode analisa yang digunakan**  

Langkah pelaksanaan penelitian dan analisis data adalah sebagai berikut:
-	Mengumpulkan data produksi daging kambing berdasarkan kabupaten/kota di jawa barat tahun 2016-2021 yang dihasilkan oleh Dinas Ketahanan Pangan dan Peternakan yang dikeluarkan dalam periode 1 tahun sekali.
-	Membuat diagram dan menganalisisnya.
-	Menghitung rata-rata produksi daging kambing di seluruh kabupaten/kota di jawa barat periode tahun 2016-2021  

Dataset yang digunakan pada penelitian ini berisi data produksi daging kambing berdasarkan kabupaten/kota di jawa barat tahun 2016-2021. Dataset terkait topik ekonomi ini dihasilkan oleh Dinas Ketahanan Pangan dan Peternakan yang dikeluarkan dalam periode 1 tahun sekali. 


```python
'Author: hernanda472@gmail.com'

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset.csv')
df_2016 = df[df['tahun'] == 2016]
df_2017 = df[df['tahun'] == 2017]
df_2018 = df[df['tahun'] == 2018]
df_2019 = df[df['tahun'] == 2019]
df_2020 = df[df['tahun'] == 2020]
df_2021 = df[df['tahun'] == 2021]
df
```

```python
print("Gambar 1 Diagram Garis Data Produksi Daging Kambing Berdasarkan Kabupaten/Kota di Jawa Barat Tahun 2016-2021")
plt.figure(figsize=(10, 6))
plt.plot(df_2016['produksi_daging'], label='2016', marker = '.', color = 'red')
plt.plot(df_2017['produksi_daging'], label='2017', marker = '.', color = 'blue')
plt.plot(df_2018['produksi_daging'], label='2018', marker = '.', color = 'green')
plt.plot(df_2019['produksi_daging'], label='2019', marker = '.', color = 'brown')
plt.plot(df_2020['produksi_daging'], label='2020', marker = '.', color = 'slateblue')
plt.plot(df_2021['produksi_daging'], label='2021', marker = '.', color = 'magenta')
plt.title("Produksi Daging Kambing Berdasarkan Kabupaten/Kota di Jawa Barat Tahun 2016-2021", size=16)
plt.xlabel('id', size=14)
plt.ylabel('Produksi Daging Kambing (kg)', size=14)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

Berdasarkan diagram batang di atas, kita dapat melihat tampilan grafik diagram garis dengan sumbu x dan sumbu y. Sumbu x menyatakan id, sedangkan pada sumbu y menyatakan data produksi daging kambing dengan satuan kilogram.  

Pada grafik di atas tedapat grafik dengan 5 warna yang berbeda yaitu, grafik berwarna merah menyatakan tahun 2016, warna biru menyatakan tahun 2017, warna hijau menyatakan tahun 2018, warna coklat menyatakan tahun 2019, warna ungu menyatakan tahun 2020, dan warna merah muda menyatakan tahun 2020.  

Produksi daging kambing tertinggi yaitu terjadi di Kabupaten Bekasi pada tahun 2019 yang menghasilkan 4820511 kilogram produksi, sedangkan produksi daging kambing tertinggi kedua yaitu terjadi pada tahun 2016, 2017, dan 2018 yang menghasilkan produksi yang hampir sama nilainya. Adapun produksi daging kambing terendah di Provinsi Jawa Barat yaitu terjadi pada tahun 2021.

```python
print("Gambar 2 Diagram Batang Data Produksi Daging Kambing Berdasarkan Kabupaten/Kota di Jawa Barat Tahun 2016")
df_2016.plot(x="nama_kabupaten_kota", y="produksi_daging", kind='bar', figsize=(10, 6), color = 'red')
plt.title('Produksi Daging Kambing Berdasarkan Kabupaten/Kota di Jawa Barat Tahun 2016', size=16)
plt.ylabel('Produksi Daging Kambing (kg)', size=14)
plt.xlabel('Kabupaten/Kota', size=14)
```

```python
rerata = df_2016.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2016 adalah:")
print(rerata)
rerata = df_2017.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2017 adalah:")
print(rerata)
rerata = df_2018.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2018 adalah:")
print(rerata)
rerata = df_2019.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2019 adalah:")
print(rerata)
rerata = df_2020.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2020 adalah:")
print(rerata)
rerata = df_2021.loc[:, 'produksi_daging':].mean()
print("Rata-rata dari produksi daging kambing (kg) di tahun 2021 adalah:")
print(rerata)
```
